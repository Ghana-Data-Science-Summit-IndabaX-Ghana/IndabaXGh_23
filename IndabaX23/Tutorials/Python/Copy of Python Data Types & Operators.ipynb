##**3.2 Storing data: Variables**

Variables are used to **store** data (also known as **value** in programming) of any sort. 

You can think of them as a depot where you put data inside and each depot has its unique name:

<img src= "https://github.com/birkfl/TReND/blob/main/Python_Workshop2022/img/variable1.svg?raw=true" height ="300px">

- **Name** of the variable: The unique name that identifies this storage space.
- **Value** of the variable: Data that is stored inside this space.
- **Type** of the variable: Data type that is stored inside this space.

**Data types will be part of the next section!**




### Creating variables: Assignments
Variables are created with an assignment statement. The syntax (kind of grammar) of variable assignment is:

```variable_name = a value``` (or an expression e.g., ```1+1```)

````python
# Define variables "year", "language", "coding minutes"
year = 2022           # year
language = "Python"   # Programming language
coding_minutes = 1550 # minutes of coding 
````

<img src= "https://github.com/birkfl/TReND/blob/main/Python_Workshop2022/img/variable_assign.svg?raw=true" height ="300px">


We just *defined* variables by assigning values to each of them. Now we can use them.

How many **hours** did we spend **coding** this year?

<img src= "https://github.com/birkfl/TReND/blob/main/Python_Workshop2022/img/variable_divide.svg?raw=true" height ="300px">



````python
# How many hours did I spend with coding in the year 2022? Print the answer!
coding_hours = coding_minutes / 60
print("Hours of coding in 2022:", coding_hours)
````

### **Changing variables**
When you use variables in other assignments or operations, they don't change!


To change variables, we need to assign them again.

````python
# (save coding_days as new variable and print it together with coding_hours)
# coding_hours / 24
coding_days = coding_hours / 24
print("coding_hours is:",coding_hours)
print("coding_days is:",coding_days)
````

Suggestion: Variable names should be **concise** (short but informative): 

- ```my_telephone_number``` -> ```phone_num```
- ```p_n``` -> ```phone_num```

### **Integers**


Below are some examples of integers

````python
#  Example of integer. How old are you? When have you been born?
age = 27
birthyear = 1995

print("I am", age, "years old.")
print("I was born in", birthyear)
````

### **Floats**
**Floating point number** (Float) is a way of representing a non-integer real number with a computer.


````python

# Let's calculate my speed from home to work
distance = 3 # in kms
time = 7 # in minutes
````

````python
# (store distance/(time/60) in a variable called speed. Print out the average speed from work and the variable type of  speed)
speed = distance/(time/60) # km/h
print("My average speed from home to work is: ", speed, "km/h")
print("'speed' variable's value is ", type(speed))
````

<a name="str"></a>
### **Strings**

What if we want to store characters instead of numbers? Usually we want to store a set of characters instead of storing each character 1 by 1 (like instead of storing - n, a, m, e - we want to store them together as -name-). This is called a string of characters. That's why the type is also called **string**.

- We should use **double quoutes ```" __ "``` or single quotes ```' __ '```** when limiting a string.



````python
# Some string examples
name = "Deborah"
country = 'Ghana'

print(name, "is from", country)
````

#### Adding blank space (a.k.a. whitespace)
- Use ```\n``` to add a new line (like pressing enter)
- Use ```\t``` for adding a tab



````python
#  Tab example
two_line_string = '\tHello  world \n A new line'
print(two_line_string)
````

#### **Concatenating (combining)**
Sometimes we want to combine strings. There are many ways of doing it. 

A fast and easy-to-read way is using the **plus** **```+```** **operator**.

````python
# The plus operator to combine strings

food = "gob3"
name = "Deborah"
surname = "Adams" 


favourite_food = name + surname + "loves" +  food # Just joining them together
favourite_food_nice = name + ' ' + surname + ' ' + "loves" + ' ' + food # Joining with a space


print(favourite_food)
print(favourite_food_nice)
````

#### **String formatting**
An easier way of concatenating is with the **f strings** (formatted string literals).



````python
# Live code - Concatenate with f string 
name = "Allan"
surname = "Mumuni"
age = 22  


print(f'My name is {name}, my surname is {surname}, my age is {age}')
````

## **4.1 Operators**

[Operators](https://www.tutorialspoint.com/python/python_basic_operators.htm) are the constructs which can manipulate the value of **operands**.

In programming, an **operand** is a term used to describe any object that is capable of being manipulated. A basic example of an operand is a variable declared. For example, in "4 + 5" the "4" and "5" are the operands and the plus symbol is the **operator**.

**Types of Operator:**
Python language supports the following types of operators.

- Arithmetic Operators *(introduced in Tutorial 1)*
- **Relational or Comparison Operators** *(introduced now)*
- **Assignment Operators** *(introduced now)*
- **Logical Operators** *(introduced now)*
- Bitwise Operators
- Membership Operators
- Identity Operators

## **4.2 What is a relational or comparison operator?**

Relational operators are also called "comparison operators" since they compare the values on either sides of them and decide the relation between them. 
So "Arithmetic operators" from the last tutorial (+, -, ...) perform operations between numbers and a "comparison/relational operator"  (<, >, ..) compares numbers (no operation is required).

**Assignment operators** are used in Python to assign values to variables, and automatically overwrite the previous value that was assigned to the variable. Written below are some examples of assignment operator:

### **The increment operator "+=" (add and assign)**

It adds right operand to the left operand and assign the result to left operand.

````python
# assume a bus with an initial amount of passengers
people_on_the_bus=10 
print(people_on_the_bus)
````

````python
#(at the next stop 2 people enter)
people_on_the_bus += 2
````

````python
# this increases the number of people on the bus by 2
print(people_on_the_bus)
````

### **The decrement operator "-=" (substract and assign)**
It subtracts right operand from the left operand and assign the result to left operand.

````python
# at the next bus stop 3 people leave the bus

people_on_the_bus -= 3
````

````python
# this decreases the number of people on the bus by 3
print(people_on_the_bus)
````

## **5.1 Conditional statement (`if`-`else`)**

### What is a conditional statement?

***Remember the following operators***

A [conditional statement](https://www.tutorialspoint.com/python/python_loops.htm) in Python allows to perform different computations or actions depending on its logical conditions such as
- "=="  (equal)
- "!="  (not equal)
- ">"   (greater than)
- "<"   (smaller than)
- ">="  (greater than or equal to)
- "<="  (smaller than or equal to)

 Conditional statements are handled by `if` statements in Python.

### **5.2 What is Python `if` Statement?**

An [`if` conditional statement](https://www.guru99.com/if-loop-python-conditional-structures.html#1) is used for decision-making operations. It contains a body of code which runs only when the condition given in the if statement is true. If the condition is false, then the optional else statement runs which contains some code for the else condition. When you want to justify one condition while the other condition is not true, then you use Python if else statement.

#### **Python `if` synthax**



`if`  statement/condition `:`    
* action 1
* action 2   

````python
#Define variables
# A product changed price:

old_price = 30;
new_price=60;
````

````python
if new_price > old_price:
  print("wooow Ghana is expensive now")
````

What will happen if we change the new_price to 24?


### **`if`...`else` flowchart**

![List](https://sorelletoukam.files.wordpress.com/2021/07/cmr11.png)

**What happens when `if` condition is not met?**


Let's change the variables of the above and see what's going to happen if the condition **if** is not met!


The `else` condition is usually used when you have to judge one statement on the basis of other. If one condition goes wrong, then there should be another condition that should justify the statement or logic.

````python

old_price = 30;
new_price=24;

if new_price > old_price:
  print("wooow Ghana is expensive now")

else:
  print("wooow it is amazing to live in this country !!!")
````

````python
#make both prices the same

old_price = 24;
new_price=24;

if new_price > old_price:
  print("wooow Ghana is expensive now")

else:
  print("wooow it is amazing to live in this country !!!")
````

**oops**!....Now both `old_price` and `new_price` over here are the same ( the price of the product didn't change) still it prints out `wooow it is amazing to live in this country !!!`.

### **When  the `else` condition does not work ...**

There might be many instances when your `else` condition will not give you the logical output but the codw will not return any errors. This happens when 
- **Script correct but conditions are vague**
- **Script correct, but the condition doesn't exists.** 

To avoid this kind of problems, we should think the problem before starting coding and we should be able to use all tools you can ( here: **if, else, elif** ).

In the EXAMPLE above, the script was correct (it run without a problem), **BUT** there was no condition to check if the prices were the same! **Hence, the condition didn't exist !**

**Further explanation**: 

- Both the variables (the price of the 1st product **price_1** and the price of the second product **price_2** are same ) and the program output is : **The new price of the product is less than the old price** .(if condition in Python), and if it fails, then it prints out the second condition (else condition) as default. However,the printed message contains a logical error! 


# Elif : How to use it #

     actions

### **Python Nested if Statement!**
### Basic structure:

   - if **condition_1** :      
     - if **condition_2**:     
           - **action for condition_2**
   - elif **condition_3**:
       - **action for condition_3**
   - else:                    
      - **else action if condition_1 or condition_2 are not met**



We can have a **if...elif...else** statement inside another **if...elif...else** statement. This is called **nesting** in computer programming.

Any number of these statements can be nested inside one another. **Indentation** is the only way to figure out the level of nesting. They can get confusing, so they must be avoided unless necessary.
