<a name="types"></a>
### **The Core Types of  Machine Learning**

* Supervised learning

* Unsupervised learning

* Reinforcement learning


#### 1. **Supervised learning**

Supervised learning is a machine learning type where  a model learns mapping between input variables and a target variable.

It is an applications in which the training data comprises examples of the input vectors along with their corresponding target vectors. 

In supervised learning,algorithms are fit on training data comprised of inputs and outputs used to make predictions on test sets where only the inputs are provided and the outputs from the model are compared to the withheld target variables and used to estimate the accuracy of the model.
    
There are two main types of supervised learning problems: they are classification that involves predicting a class label and regression that involves predicting a numerical value.

**Classification**: Supervised learning problem that involves predicting a class label.

Example:
* Loan worthiness prediction


**Regression**: Supervised learning problem that involves predicting a numerical label.

Example
* House price prediction

Both classification and regression problems may have one or more input variables and input variables may be any data type, such as numerical or categorical.

#### 2. **Unsupervised learning**

Unsupervised learning refers to the learning type where a model describes or extracts relationships in data. It looks for patterns in unlabeled data.

Unsupervised learning operates on only input data without outputs or target variables. This implies that unsupervised learning does not have a teacher correcting the model, as in the case of supervised learning.

Basically, unsupervised learning is used for:

* Clustering: Unsupervised learning problem that involves finding groups in data.
Example: Customer segmentation

* Density Estimation: Unsupervised learning problem that involves summarizing the distribution of data.



#### 3. **Reinforcement Learnng**
This is the learning type which is based on reward and punishment system. It trains machines through trial and error to do the right thing by rewarding it. In Reinforcement Learning, an agent learns what to do so as to maximize a numerical reward signal. 
The learner is not told which actions to take, but instead must discover which actions yield the most reward by trying them.
In this type of learning there is no training dataset, rather there is  a goal or set of goals that an agent is required to achieve, actions they may perform, and feedback about performance toward the goal.
Common application of Reinforcement Learnng includes training models to play games, and self-driving cars.

<a name="pipeline"></a>
### **Machine Learning Pipeline (Step by step process to build a Machine Learning project)**





1. Problem statement (Business understanding)
   + Clearly state the busisness objective or the problem statement
   + Check if the problem can be solved with a Machine Learning technique 

2. Identification of the appropriate Machine Learning technique based on the problem
   + Choose the appropriate Machine Learning technique to solve the problem.It could be regression, classification, clustering, recommendation system, etc.
3. Data Collection
   + Collect the needed data for the solution
4. Data exploration and data cleaning
   + Explore the data to get a better undersatnding and make assumptions
   + Deal with data contraints appropriately. eg. missing values, type errors, duplicates etc.
5. Data preparation for modelling
   + Feature engineering and Feature selection
   + Do feature encoding where necessary
   + Data scaling where necessary
   + Separate your data into input (independent) variables and target(dependent) variable where necessary.
   + Split your data into train set, validation set and test set.
6. Model building
   + Build a base model with a simple algorithm to have a fair knowledge of how your models will perform.
   + Build more models with more advanced algorithms

7. Model improvement
   + Do hyperparameter tuning using pipelines and cross validation. This will help you correct overfitting or underfitting and make your model generalize well.
   + After hyperparameter tuning, you can train the models on the whole train set again.
   
8. Model evaluation
  + Evaluate your models with an  appropriate metric. In a classification case consider whether the data is balanaced or not before choosing a metric.
  + If the models' outcome does not meet your expectation, you can repeat the previous steps and make changes.

9. Model selection
   + Based on your objectives and the evaluation metrics choose the best model for your data.
10. Model Deployment
   + Deploy your chosen model.
11. Model monitoring
   + Monitor your model to know when it drifts
   + Re-train your model when necessary

### **Regression**

## **What is regression**?
Regression is a statistical method used to determine the strength and nature of the relationship between a dependent variable and an independent variable (s). In simple terms regression finds correlations between dependent and independent variables and allows us to make predictions.

There are different types regression but linear regression is the most common one. Linear regression establishes the linear relationship between two variables based on a line of best fit. Linear regression is usaully graphically depicted using a straight line with the slope defining how the change in one variable impacts a change in the other as shown.

<img src="https://miro.medium.com/v2/resize:fit:720/format:webp/1*G1Y_-X14q2xMVHlUuaUUdA.png">

[Source](https://towardsdatascience.com/introduction-to-linear-regression-in-python-c12a072bedf0)

The y-intercept of a linear regression relationship represents the value of one variable when the value of the other is zero. 

When there is only one independent variable, it termed as simple linear regression

Y = a + bX + e

Multiple linear regression on the hand has more independent variable.

Y=a + b1X1 + b2X2 + b3X3+...+btXt + e

where:

Y=The dependent variable you are trying to predictor explain

X=The explanatory (independent) variable(s) you are using to predict or associate with Y

a = The y-intercept

b = (beta coefficient) is the slope of the explanatory variable(s)

e = The regression residual or error term

There are also Non-linear regression models which are far more complex

<a name="ols"></a>
### **Ordinary Least Square (OLS**

Different tehniques are used in estimating coefficients of linear regression equations which describe the relationship between one or more independent variables and a dependent variable.

Ordinary Least Squares (OLS) is a common technique and seeks to minimize minimum sum of squared errors (SSE). It is an optimization strategy that allows you to find a straight line that is as close as possible to your data points in a linear regression model. More complext techniques include Maximum likelihood, and Generalized method of moments.

<a name="assumptions"></a>
### **Assumptions of Linear regression**

* Linear relationship
* No or little multicollinearity
* No auto-correlation
* Homoscedasticity
* Normality: The residuals of the model are normally distributed.
